```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
online=pd.read_excel('Online retail.xlsx',header=None)
online.head()
```

# Data Preprocessing:

```python
df=pd.get_dummies(online,dtype='int')
df.head()
```

#Association Rule Mining:

#Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift",min_threshold=1)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.01,num_itemsets=10)
rules
```

#Analysis and Interpretation:

```python
rules = rules[(rules['support'] >= 0.01)&
              (rules['confidence'] >= 0.3) &
              (rules['lift'] >= 3)]
```

```python
rules
```

```python
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

#What is lift and why is it important in Association rules?

* It measures how much more often items X and Y occur together than expected if they were statistically independent. A lift > 1 indicates a positive association.




#What is support and Confidence. How do you calculate them?


* Support is the frequency of itemset in the dataset we calculate
  - Support(A) = Transactions with A/Total Transaction

* Confidence is Likelihood that Y is purchased when X is purchased we calculade
  - Confidence(A -> B) =Support(A∪B)/Support(A)

                





What are some limitations or challenges of Association rules mining?

* Can result in too many rules (many not meaningful)

* Requires significant computational power with large datasets

* Doesn't capture sequential patterns

* Rare items may be ignored due to low support
