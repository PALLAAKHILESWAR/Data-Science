#State the Hypotheses
We need to check there is significant difference in the hypothesis or not by chicking Ho and  Ha

```python
data = {"Satisfaction": ["Very Satisfied", "Satisfied", "Neutral", "Unsatisfied", "Very Unsatisfied", "Total"],
    "Smart Thermostat": [50, 80, 60, 30, 20, 240],
    "Smart Light": [70, 100, 90, 50, 50, 360],
    "Total": [120, 180, 150, 80, 70, 600]
}
```

```python
dataframe = pd.DataFrame(data)
```

```python
dataframe.iloc[0:5,1:3]
```

```python
crosstable=pd.crosstab(dataframe["Satisfaction"], dataframe["Smart Thermostat"])
```

```python
crosstable
```

# Compute the Chi-Square Statistic

```python
from scipy.stats import chi2_contingency
chi,p,dof,expected=chi2_contingency(dataframe.iloc[0:5,1:3])
```

```python
print(chi,p,dof)
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(dataframe.iloc[0:5,1:3])
```
